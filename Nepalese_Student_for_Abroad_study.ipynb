```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the CSV
df = pd.read_csv("Nepali_Students_Abroad_2018_2024.csv")
df.head()
```

```python
# Remove "~" and commas, convert to numeric
for col in df.columns[1:]:
    df[col] = df[col].str.replace("~", "").str.replace(",", "").astype(float)

df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["Year"], df["Total_Students"], marker='o', color='blue')
plt.title("Total Nepali Students Going Abroad (2018â€“2024)")
plt.xlabel("Year")
plt.ylabel("Total Students")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_countries = df.set_index("Year")[["Australia", "Japan", "India", "USA", "Canada"]]

# Stacked bar
df_countries.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="Set3")
plt.title("Nepali Students Abroad by Country (2018â€“2024)")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
totals = df_countries.sum()

plt.figure(figsize=(6, 6))
plt.pie(totals, labels=totals.index, autopct='%1.1f%%', startangle=140)
plt.title("Total Students by Country (2018â€“2024)")
plt.tight_layout()
plt.show()
```

```python
print("Maximum Total Students in a Year:", int(np.max(df["Total_Students"])))
print("Minimum Total Students in a Year:", int(np.min(df["Total_Students"])))
print("Average Total Students per Year:", int(np.mean(df["Total_Students"])))
```
